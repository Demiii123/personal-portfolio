```python
#Step 3: High Charges Analysis

query = """select top 10 customer_id as 'Customer ID', year as Year, charges as Charges
from hospitalisation_details where charges>700;"""
df = pd.read_sql(query, conn)
df
```

```python
#Step 4: High BMI Patients Analysis

query = """SELECT top 10 n.name as Name, h.year as Year, h.charges as Charges
FROM hospitalisation_details AS h
JOIN medical_examinations AS m ON h.customer_id = m.customer_id
JOIN names AS n ON n.customer_id = m.customer_id
WHERE m.BMI > 35
"""
df = pd.read_sql(query, conn)
df
```

```python
#Step 5: Customers with Major Surgeries

query = """SELECT top 10 n.customer_id as 'Customer ID', n.name as 'Customer Name'
FROM names AS n
JOIN medical_examinations AS m ON n.customer_id = m.customer_id
WHERE m.numberofmajorsurgeries >= 1
"""
df = pd.read_sql(query, conn)
df
```

```python
#Step 6: Average Charges by Hospital Tier in 2000

query = """select hospital_tier as 'Hospital Type', avg(charges) as 'Average Charges' from hospitalisation_details
where year = 2000
group by hospital_tier
"""

df = pd.read_sql(query, conn)
df
```

```python
#Step 7: Smoking Patients with Transplants Analysis

query = """
SELECT 
    m.customer_id as 'Customer ID', 
    m.BMI, 
    h.charges as Charges
FROM 
    medical_examinations AS m
JOIN 
    hospitalisation_details AS h 
    ON h.customer_id = m.customer_id
WHERE 
    smoker = 1 AND any_transplant = 1
"""

df = pd.read_sql(query, conn)
df.head()
```

```python
#Step 8: Patients with Major Surgeries or Cancer History

query = """SELECT TOP 10 n.name AS Name 
FROM names AS n
JOIN medical_examinations AS m ON m.customer_id = n.customer_id
WHERE cancer_history = 1 OR numberofmajorsurgeries >= 2
"""

df = pd.read_sql(query, conn)
df
```

```python
#Step 9: Customer with Most Major Surgeries

query = """select n.customer_id as 'Customer ID', n.name as Name 
from  names as n
JOIN medical_examinations as m ON m.customer_id = n.customer_id
order by numberofmajorsurgeries desc
"""

df = pd.read_sql(query, conn)
df
```

```python
#Step 10: Customers with Major Surgeries and City Tiers

query = """SELECT n.customer_id as 'Customer ID', n.name as Name, h.city_tier as City
FROM hospitalisation_details AS h
JOIN medical_examinations AS m ON h.customer_id = m.customer_id
JOIN names AS n ON n.customer_id = m.customer_id
where numberofmajorsurgeries > 0
"""

df = pd.read_sql(query, conn)
df
```

```python
#Step 11: Average BMI by City Tier in 1995

query = """select h.city_tier as City, AVG(m.BMI) as 'Average BMI'
FROM hospitalisation_details AS h
JOIN medical_examinations AS m ON h.customer_id = m.customer_id
where year = 1995
group by h.city_tier
"""

df = pd.read_sql(query, conn)
df
```

```python
#Step 12: High BMI Customers with Health Issues

query = """SELECT top 10 n.customer_id as 'Customer ID', n.name as Name, h.charges as Charges
FROM hospitalisation_details AS h
JOIN medical_examinations AS m ON h.customer_id = m.customer_id
JOIN names AS n ON n.customer_id = m.customer_id
where health_issues = 1 and m.BMI > 30
"""

df = pd.read_sql(query, conn)
df
```

```python
#Step 13: Customers with Highest Charges and City Tier by Year

query = """SELECT h.year as Year, n.name as Name, h.city_tier as City, max(h.charges) as 'Max Charges'
FROM hospitalisation_details AS h
JOIN names AS n ON n.customer_id = h.customer_id
group by h.year, n.name, h.city_tier
Having max(h.charges) = (select max(charges) from hospitalisation_details where year = h.year)
"""

df = pd.read_sql(query, conn)
df.head(10)
```

```python
#Step 14: Top 3 Customers with Highest Average Yearly Charges

query = """with Yearlycharges as ( 
    select customer_id, avg(charges) as avg_yearly_Charges 
    from hospitalisation_details group by customer_id, year)
Select n.name as Name, y.avg_yearly_Charges as 'Average Yearly Charges' from names n
JOIN Yearlycharges as y ON y.customer_id = n.customer_id
order by y.avg_yearly_Charges desc
"""

df = pd.read_sql(query, conn)
df.head(10)
```

```python
#Step 15: Ranking Customers by Total Charges

query = """SELECT top 10 n.name as Name, sum(h.charges) as 'Total Charges', 
RANK() OVER (ORDER BY SUM(h.charges) desc) AS 'Charges Rank'
FROM hospitalisation_details AS h
JOIN names AS n ON n.customer_id = h.customer_id
group by n.name
order by 'Charges Rank' asc
"""

df = pd.read_sql(query, conn)
df.head(10)
```

```python
#Step 16: Identifying Peak Year for Hospitalizations

query = """WITH YearlyHospitalizations AS (
    SELECT
        year,
        COUNT(*) AS num_hospitalizations
    FROM
        hospitalisation_details
    GROUP BY
        year
)
SELECT
    year,
    num_hospitalizations
FROM
    YearlyHospitalizations
WHERE
    num_hospitalizations = (SELECT MAX(num_hospitalizations) FROM YearlyHospitalizations)
"""

df = pd.read_sql(query, conn)
df
```

```python
#Step 17 :Yearly Trend of Smoking Patients

query = """SELECT h.year, COUNT(*) AS 'Smoker Count'
FROM hospitalisation_details h
JOIN medical_examinations m ON h.customer_id = m.customer_id
WHERE m.smoker = 1
GROUP BY h.year
ORDER BY h.year
"""

df = pd.read_sql(query, conn)
df.head(10)
```

```python
#Step 18 :Repeat Hospitalized Customers

query = """SELECT top 10 customer_id, COUNT(*) AS visit_count
FROM hospitalisation_details
GROUP BY customer_id
HAVING COUNT(*) > 1
ORDER BY visit_count DESC
"""

df = pd.read_sql(query, conn)
# df.head(10)

print("Repeat Customers:", len(df))

# It means we have no repeat customers
```

```python
# Step 19 :BMI Category vs Average Charges

query = """
SELECT
  CASE
    WHEN m.BMI < 18.5 THEN 'Underweight'
    WHEN m.BMI BETWEEN 18.5 AND 24.9 THEN 'Normal'
    WHEN m.BMI BETWEEN 25 AND 29.9 THEN 'Overweight'
    WHEN m.BMI BETWEEN 30 AND 34.9 THEN 'Obese I'
    WHEN m.BMI BETWEEN 35 AND 39.9 THEN 'Obese II'
    ELSE 'Obese III'
  END AS BMI_Category,
  AVG(h.charges) AS 'Avg Charges'
FROM hospitalisation_details h
JOIN medical_examinations m ON h.customer_id = m.customer_id
GROUP BY
  CASE
    WHEN m.BMI < 18.5 THEN 'Underweight'
    WHEN m.BMI BETWEEN 18.5 AND 24.9 THEN 'Normal'
    WHEN m.BMI BETWEEN 25 AND 29.9 THEN 'Overweight'
    WHEN m.BMI BETWEEN 30 AND 34.9 THEN 'Obese I'
    WHEN m.BMI BETWEEN 35 AND 39.9 THEN 'Obese II'
    ELSE 'Obese III' end
ORDER BY 'Avg Charges' DESC

"""

df = pd.read_sql(query, conn)
df.head(10)

# It means we have no repeat customers
```

```python
sns.set(style="whitegrid")

# Rename column if it appears as 'Avg Charges' with quotes
df.columns = [col.strip("'") for col in df.columns]

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='BMI_Category', y='Avg Charges', palette='magma')

# Titles and labels
plt.title('BMI Category vs Average Charges', fontsize=16)
plt.xlabel('BMI Category', fontsize=12)
plt.ylabel('Average Charges', fontsize=12)

# Rotate x-axis labels
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```
